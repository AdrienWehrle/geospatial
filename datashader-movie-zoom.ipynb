````python
import os
import numpy as np
import tqdm
import datashader as ds
import datashader.transfer_functions as tf
from datashader.colors import inferno, viridis
from datashader.transfer_functions import set_background
from colorcet import palette
import pandas as pd
import matplotlib.pyplot as plt
import ffmpeg

import holoviews as hv
from colorcet import fire
import dask.dataframe as dd
from colorcet import fire
from holoviews.operation.datashader import datashade
hv.extension('bokeh', 'matplotlib')
from holoviews import opts

# References
# https://casyfill.github.io/posts/2016/08/big_data_mapping/
# https://aetperf.github.io/2020/02/13/Lunch-break,-plotting-traffic-injuries-with-datashader.html
````

# Process data

````python
df = pd.read_parquet('data/geolife_trajectories_1_3.parquet')
print(df.shape[0])
df.head()
````

# Make a video with a zooming into the city

````python
# Define parameters
width           = 800
height_to_width = 1
init_frames     = 50
zoom_frames     = 150
final_frames    = 100

# Define zoom rectangles 
start_lat_min = 35
start_lat_max = 45
start_lon_min = 115
start_lon_max = 120
  
final_lat_min = 39.970
final_lat_max = 39.985
final_lon_min = 116.326
final_lon_max = 116.340

y_range    = np.linspace((start_lat_min, start_lat_max), (final_lat_min, final_lat_max), zoom_frames)
x_range    = np.linspace((start_lon_min, start_lon_max), (final_lon_min, final_lon_max), zoom_frames)
````

````python
# Static view of all of all the data
cvs = ds.Canvas(plot_width=width, plot_height=int(width*height_to_width), x_range=x_range[0], y_range=y_range[0])
agg = cvs.points(df, 'lon', 'lat')
img = tf.shade(agg,  cmap=palette["fire"], how='log')
img = set_background(img, 'black')

# Generate a bunch of static images for the full map
frame = 0
for itt in tqdm.tqdm(range(init_frames)):
    filename = f"data/datashader_movie_zoom_frame_{frame:04}_intial"
    ds.utils.export_image(img, filename, fmt=".png")
    frame += 1

# Zoom towards the target
for fig_no, (x_rng, y_rng) in tqdm.tqdm(enumerate(zip(x_range, y_range)), total=len(list(zip(x_range, y_range)))):
    cvs = ds.Canvas(plot_width=width, plot_height=int(width*height_to_width), x_range=x_rng, y_range=y_rng)
    agg = cvs.points(df, 'lon', 'lat')

    img = tf.shade(agg,  cmap=palette["fire"], how='log')
    img = set_background(img, 'black')
    
    filename = f"data/datashader_movie_zoom_frame_{frame:04}_zoom"
    ds.utils.export_image(img, filename, fmt=".png")
    frame += 1

# Write the last image a couple of extra times to stall the video on the final frame
for fig_no_extra in tqdm.tqdm(range(final_frames)):
    filename = f"data/datashader_movie_zoom_frame_{frame:04}_final"
    ds.utils.export_image(img, filename, fmt=".png")
    frame += 1
````

# Create a movie from the images

First you need to join the images into a movie using the ffmpeg tool in Linux

```bash
ffmpeg -y -r 25 -f image2 -pattern_type glob -i 'data/datashader_movie_zoom_*.png' -vcodec libx264  -crf 25 -pix_fmt yuv420p data/datashader_movie_zoom.mp4
```
This will create a mp4 video in the data folder with 25 frames per second.
