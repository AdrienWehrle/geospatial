```python
import pandas as pd
```

# Read the dataset

```python
df = pd.read_csv('data/0003827-190918142434337.zip', compression='zip', sep='\t')
print(df.shape[0])
df.head()
```

## Basic clean up

```python
# Limit the dataset, only use some of the columns
df = df[['species','decimalLatitude', 'decimalLongitude', 'eventDate']]

# Make sure that the datetime string is correctly read
df['eventDate'] = pd.to_datetime(df['eventDate'])

# Remove incomplete records
df = df[
    (~df['decimalLatitude'].isna())
    & (~df['decimalLongitude'].isna())
    & (~df['species'].isna())
    & (~df['eventDate'].isna())
]

# Rename the columns for shorter code
df.rename(columns={'decimalLatitude': 'lat', 'decimalLongitude': 'lon'}, inplace=True)

print(df.shape[0])
df.head()
```
